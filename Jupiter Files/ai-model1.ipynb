```python
fig,axes = plt.subplots(nrows=1,ncols=3,figsize=(16,6))


axes[0].plot(df_copy['Number_of_Emloyees'],df_copy['Store_Sales'],'o')
axes[0].set_ylabel("Sales")
axes[0].set_title("Number_of_Emloyees")


axes[1].plot(df_copy['Number_of_products'], df_copy['Store_Sales'],'o')
axes[1].set_ylabel("Sales")
axes[1].set_title("Number_of_products")


axes[2].plot(df_copy['Area_of_Store'], df_copy['Store_Sales'],'o')
axes[2].set_ylabel("Sales")
axes[2].set_title("Area_of_Store")


plt.tight_layout()
plt.show()


```

```python
sns.pairplot(df_copy, diag_kind='hist')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X = df_copy.drop('Store_Sales', axis=1)
y = df_copy['Store_Sales']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=101)
```

```python
print(len(df_copy))
print(len(X_train))
print(len(y_train))
print(len(X_test))
print(len(y_test))
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```
