```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.arange(0, 10)
y = 2 * x
```

```python
print(x)
print(y)
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)
plt.title('Example')
plt.xlabel('X DATA')
plt.ylabel('Y DATA')
plt.xlim(0,6)
plt.ylim(0,15)
plt.show()
```

```python
plt.plot(x, y)
plt.title('Example')
plt.xlabel('X DATA')
plt.ylabel('Y DATA')
plt.xlim(0,6)
plt.ylim(0,15)
plt.savefig('plot_example.jpg')
plt.show()
```
