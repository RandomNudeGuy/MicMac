```python
import matplotlib.pyplot as plt
import numpy as mp
import pandas as pd
```

```python
data = {
'Manufacturer': ['Toyota', 'Ford', 'Ford', 'Audi', 'BMW', 'Toyota', 'Audi', 'BMW', 'Nissan', 'Nissan'],
'Type': ['Sedan', 'Truck', 'SUV', 'Sedan', 'Sedan', 'Truck', 'SUV', 'Truck', 'SUV', 'Sedan'],
'Fuel_Efficiency': [34, 22, 24, 27, 28, 24, 26, 21, 23, 35],
'Horsepower': [180, 320, 260, 220, 240, 290, 210, 310, 250, 200],
'Price': [30000, 40000, 35000, 45000, 50000, 32000, 47000, 48000, 33000, 31000]
}
```

2. Plot a line chart representing the 'Horsepower' vs 'Price' relationship for all manufacturers. Use title, axis labels and axis limitation. Make your line chart dashed style (--), in red color and add the actual data points on the chart itself (marker).

```python
hp = data['Horsepower']
price = data['Price']
plt.title('HP VS. Price')
plt.xlabel('HP')
plt.ylabel('Horsepower')
plt.plot(hp, price, ls='--', marker='o', color='r')
plt.show()
```

Calculate the average price according to each car 'Type', Plot a line chart to show the average price for all the different car types. Add legend to differentiate between the different lines in your chart. Add title and axis labels.

```python
data_df = pd.DataFrame(data)
```

```python
type_price = data_df.groupby('Type')['Price'].mean()
```

```python
typee = type_price.index
price_type = type_price.values
plt.title('Type VS. Price')
plt.xlabel('Type')
plt.ylabel('Price')
plt.plot(typee, price_type, ls='--', marker='o', color='r')
plt.show()
```

Calculate the average 'Horsepower' according to each car 'Type', Create a bar chart that represents the average Horsepower for each vehicle Type. Add legend to differentiate between the different bars in your chart. Add title and axis labels.

```python
type_horsepower = data_df.groupby('Type')['Horsepower'].mean()
```

```python
horsepower = type_horsepower.index
price_horsepower = type_horsepower.values
plt.title('Horsepower VS. Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.bar(horsepower, price_horsepower)
plt.show()
```

Plot a histogram to visualize the distribution of Fuel Efficiency, use 5 bins in your histogram chart. Add title and axis labels.

```python
fuel_e = data['Fuel_Efficiency']
plt.title('Fuel Efficiency')
plt.xlabel('Fuel Efficiency')
plt.ylabel('Amount')
plt.hist(fuel_e, bins=5)
plt.show()
```

Create a scatter plot to explore the relationship between Horsepower and Price. Determine according to the chart if there is any correlation between them. Add title, and axis labels.

```python
hp = data['Horsepower']
price = data['Price']
plt.title('HP VS. Price')
plt.xlabel('HP')
plt.ylabel('Price')
plt.scatter(hp, price)
plt.show()
#there is no correlation 
```

Use the chart you created in exercises 2, 4, 5, 6 and combine them to one chart using subplot (2 X 2 chart grid). Output your final chart to a jpeg file.

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 10))

ax[0, 0].plot(typee, price_type, ls='--', marker='o', color='r')
ax[0, 0].set_title('HP VS. Price')

ax[0, 1].scatter(hp, price)
ax[0, 1].set_title('HP VS. Price')

ax[1, 0].bar(horsepower, price_horsepower)
ax[1, 0].set_title('Horsepower VS. Price')

ax[1, 1].hist(fuel_e, bins=5)
ax[1, 1].set_title('Fuel Efficiency')

fig.savefig('hw.jpeg')
plt.show()
```
