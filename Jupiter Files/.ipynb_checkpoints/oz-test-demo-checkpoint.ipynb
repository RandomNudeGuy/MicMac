```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
students_df = pd.read_csv('/Users/mcrandom/Downloads/students (1) (1).csv')
courses_df = pd.read_csv('/Users/mcrandom/Downloads/courses (1) (1).csv')
teachers_df = pd.read_csv('/Users/mcrandom/Downloads/teachers (1) (1).csv')
course_student_df = pd.read_csv('/Users/mcrandom/Downloads/studnet_course (1) (1).csv')
```

```python
students_df_copy = students_df.copy()
students_df_copy
##copy the df
```

```python
students_df_copy_dupes = students_df_copy[students_df_copy.duplicated(subset='email', keep=False)]
students_df_copy_dupes
##find the dupes by email and keep beauce duplicated delets first itiration 
```

```python
students_df_copy_dupes['NaN Count'] = students_df_copy_dupes.count(axis=1)
#create new collum to count Nan by counting how many values each row has (axis=1). less than the number of collums, has Nan
```

```python
final_students_df_copy_dupes = students_df_copy_dupes.sort_values(by=['email', 'NaN Count'], ascending=False)
final_students_df_copy_dupes
#sorting by email and nan count for better view. "ascending=False" for high to low on the NaN count
```

```python
kept = final_students_df_copy_dupes.drop_duplicates('email')
kept
#saving whats left of removing the dupes from the "final_students_df_copy_dupes" we sorted for later use
```

```python
removed = final_students_df_copy_dupes[~final_students_df_copy_dupes["id"].isin(kept["id"])]
removed
#saving what is NOT (in) kept  from final_students_df_copy_dupes (the dupes we dont want) for later use
```

```python
students_df_copy = students_df_copy[~students_df_copy["id"].isin(removed["id"])]
#saving what is NOT (in) removed from students_df_copy (not the dupes, that we do want to keep)
students_df_copy_dupes = students_df_copy[students_df_copy.duplicated(subset='email', keep=False)]
students_df_copy_dupes
#check for dupes again
```

```python
course_student_df_copy = course_student_df.copy()
course_student_df_copy = course_student_df_copy[~course_student_df_copy['student_id'].isin(removed['id'])]
#creating a copy of course_student_df and saving what is NOT (in) removed from course_student_df_copy by student_id (not the dupes, that we do want to keep)
```
