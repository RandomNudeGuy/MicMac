```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import scipy
```

```python
df = pd.read_csv('/Users/ben.meir/Downloads/hearing_test (3).csv')

df
```

```python
sns.countplot(data=df, x=df['test_result'])
plt.show
```

```python
sns.boxplot(data=df, x='test_result', y='age')
plt.show()
```

```python
sns.scatterplot(data=df, x='age', y='physical_score', hue='test_result', alpha=0.5)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression


X = df.drop('test_result', axis=1)
y = df['test_result']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)

scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = scaler.transform(X_train)
scaled_X_test = scaler.transform(X_test)

logistic_model = LogisticRegression()
logistic_model.fit(scaled_X_train, y_train)
```

```python
y_predictions = logistic_model.predict(scaled_X_test)
```

```python
y_predictions
```

```python
logistic_model.coef_
```

```python
y_preb_predictions = logistic_model.predict_proba(scaled_X_test)
y_preb_predictions
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = logistic_model.predict(scaled_X_test)
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
y_test.value_counts()
```

```python
cm = confusion_matrix(y_test, y_pred)
cmap = plt.cm.viridis

sns.heatmap(cm, annot=True, fmt="d", cmap=cmap)

plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score

recall = recall_score(y_test, y_pred)
recall
```

```python
precision = precision_score(y_test, y_pred)
precision
```

```python
f1_score = 2 * (recall * precision) / (recall + precision)
f1_score
```

```python
print(classification_report(y_test, y_pred))
```
